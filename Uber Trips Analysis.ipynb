```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('uber-raw-data-sep14.csv')
df.head()
```

```python
df['Date/Time']=pd.to_datetime(df['Date/Time'])
df.dtypes
```

```python
df["Day"] = df["Date/Time"].apply(lambda x: x.day)
df["Weekday"] = df["Date/Time"].apply(lambda x: x.weekday())
df["Hour"] = df["Date/Time"].apply(lambda x: x.hour)
print(df.head())
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
sns.histplot(data=df)
plt.show()
```

```python
#By examining the daily travels, we may conclude that Uber usage increases during the week and declines on the weekends. Let's now examine the Uber journeys based on the hours.
```

```python
sns.histplot(data=df, x='Day')
plt.show()
```

```python
# According to hourly data, Uber trips start to decline after midnight and then begin to rise after 5 am. They continue to rise until 6 pm, which is the busiest hour for Uber, after which they begin to decline.
```

```python
sns.histplot(data=df,x='Hour')
plt.show()
```

```python
# Since Sundays have more Uber rides than Saturdays do in the preceding graph, we can infer that individuals use Uber for other purposes other just getting to work. Uber journeys are at their lowest volume on Saturdays and at their highest volume on Mondays.
```

```python
sns.histplot(data=df,x='Weekday')
plt.show()
```

```python
# With the information on longitude and latitude, we can also map the number of Uber rides each region in New York City:
```

```python
sns.scatterplot(data=df, x='Lon', y='Lat').set(title='Uber Trips Analysis')
plt.show()
```

```python
## So, we can examine the Uber journeys in New York City in this manner. My findings from this analysis include the following, among others:

# Uber makes the most money on Mondays.
# Less people utilize Uber on Saturdays.
# The busiest hour for Uber is about 6 p.m.
# An increase in Uber trips often starts around 5 am.
# Near the New York region of Manhattan, the majority of Uber rides start.
```
